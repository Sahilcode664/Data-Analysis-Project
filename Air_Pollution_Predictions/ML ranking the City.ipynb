```python
y=df["Rank"]
X=df.select_dtypes(["number"])
X=X.drop(columns="Rank")
#X=X.drop(columns="Life expectancy(years) (Country)")
#X=X.drop(columns="Happiness levels(Country)")
#X=X.drop(columns="Outdoor activities(City)")
#X=X.drop(columns="Sunshine hours(City)")
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test , y_train ,y_test = train_test_split(X,y,test_size = 0.3,random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import mean_squared_error,r2_score
classifier = SVC(kernel = "linear",random_state=0)
classifier.fit(X_train_scaled,y_train)
y_pred=classifier.predict(X_test_scaled)
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
classifier.predict(X_train_scaled)
```

```python
classifier.predict(X_test_scaled)
```

```python
classifier.score(X_test_scaled,y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
model=LinearRegression()
model.fit(X_train_scaled,y_train)
y_pred=model.predict(X_test_scaled)
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression 
log_reg= LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
y_pred=log_reg.predict(X_test_scaled)
r2_score(y_test,y_pred)
```
