```python
# Finding Missing Values in Dataset using Missing as a Null value in dataset
missing_values=["Missing"]
df=pd.read_csv("Laptop_data.csv",na_values=missing_values)
```

```python
null_val=df.isnull().sum()
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

Data Cleaning

```python
# Cleaning the dataset using the none value recognizations
for column in (df.columns):
    if null_val[column] > 80:
        df=df.dropna(axis=1)
df.info()
```

```python
df['Sucess'] = [1 if x >= 4  else 0 for x in df['star_rating']]
df
```

```python
a=0
b=0
for x in df['Sucess']:
    if x==0:
        a+=1
    else:
        b+=1
print("Success = ",b,"\nUnsucess = ",a)
```

```python
y=df["Sucess"]
X=df.select_dtypes(["number"])
X=X.drop(columns="Sucess")
X
```

Spliting the dataset for the Prediction 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test , y_train ,y_test = train_test_split(X,y,test_size = 0.3,random_state=21)
```



Standardization of the Value in X_train for ML Modeling




```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

Prediction on the basis of the LogisticRegression

```python
from sklearn.linear_model import LogisticRegression 
```

```python
log_reg= LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
```

```python
log_reg.predict(X_test_scaled)
```

```python
log_reg.score(X_train_scaled,y_train)
```

```python
Log_Reg_score=log_reg.score(X_test_scaled,y_test)
Log_Reg_score
```

Prediction on the basis of the SVM 

```python
from sklearn.svm import SVC 
classifier = SVC(kernel = "linear",random_state=0)
classifier.fit(X_train_scaled,y_train)
```

```python
classifier.predict(X_train_scaled)
```

```python
classifier.score(X_train_scaled,y_train)
```

```python
SVM_score=classifier.score(X_test_scaled, y_test)
SVM_score
```

Prediction on the basis of the KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier()
model.fit(X_train_scaled,y_train)
model.predict(X_train_scaled)
```

```python
model.score(X_train_scaled,y_train)
```

```python
KNN_score=model.score(X_test_scaled,y_test)
KNN_score
```

Result of the Model after Score

```python
data = [['Logistic Regression', Log_Reg_score], ['SVM', SVM_score], ['KNN', KNN_score]]

# Create the pandas DataFrame
df1 = pd.DataFrame(data, columns=['Model', 'Score'])

df1
```

```python
my_plot=sns.scatterplot(x="Model",y="Score",data=df1)
my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=90)
my_plot
```
